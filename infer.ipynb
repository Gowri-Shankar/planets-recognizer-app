```python
# Basic Packages
import numpy as np
import matplotlib.pyplot as plt

# Import Fast AI Library
from fastai import *
from fastai.vision import *

# Import Utility Object
from RAdam import RAdam
```

```python
model_dir = Path("../models/")
data_dir = Path("../")
test_dir = Path('../test/')
```

## Loading Pretrained Model with Weights

```python
learner = load_learner(model_dir)
```

## Inference on One Test Sample

```python
img = open_image(test_dir/'6.jpg')
img.show()
str(learner.predict(img)[0])
```

## Making Inference on All Planets in Test Set

```python
_, axs = plt.subplots(2,4, figsize=(20,8))
i = 1
for ax in axs.flatten():
    img = open_image(test_dir/f'{i}.jpg')
    img.show(ax, y=learner.predict(img)[0])
    i += 1
```
